# Practical 2: Decision Tree Classifier


### Decision Tree Classifier

- A **supervised learning algorithm** for **classification & regression**.
- Recursively splits the dataset based on **feature values** to create a tree.
- **Internal nodes** → Decision rules, **Leaf nodes** → Class labels.

#### Steps:
1. Import necessary libraries.
2. Load **Iris dataset**.
3. Split into **training & testing sets**.
4. Train a **Decision Tree Classifier**.
5. Predict **test labels**.
6. Evaluate using **accuracy & confusion matrix**.
7. **Visualize** the decision tree.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn import tree
print("Libraries Imported Successfully!")
```

```python

iris = load_iris()
X = iris.data
y = iris.target
print("Dataset Loaded Successfully!")
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
print("Data Split into Training and Testing Sets!")
```

```python

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print("Decision Tree Model Trained!")
```

```python

y_pred = clf.predict(X_test)
print("Predictions Made!")
```

```python

accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
print("Confusion Matrix Displayed!")
```

```python

plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Visualization")
plt.show()
print("Decision Tree Visualized!")
```
